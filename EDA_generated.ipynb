```python
import pandas as pd

file_path = 'final_with_ratios_synthetic_distribution.csv'
data = pd.read_csv(file_path)

data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the histograms for Age and Speed of Payment
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Age distribution
axs[0].hist(data['Age'], bins=30, color='blue', edgecolor='black')
axs[0].set_title('Age Distribution')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Frequency')

# Speed of Payment distribution
axs[1].hist(data['speed_of_payment'], bins=30, color='green', edgecolor='black')
axs[1].set_title('Speed of Payment Distribution')
axs[1].set_xlabel('Speed of Payment')
axs[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Plot histograms for key numerical variables
fig, axs = plt.subplots(3, 2, figsize=(12, 12))

# Plotting Income
axs[0, 0].hist(data['Income'], bins=30, color='blue', edgecolor='black')
axs[0, 0].set_title('Income Distribution')
axs[0, 0].set_xlabel('Income')
axs[0, 0].set_ylabel('Frequency')

# Plotting Annual Income
axs[0, 1].hist(data['Annual Income'], bins=30, color='green', edgecolor='black')
axs[0, 1].set_title('Annual Income Distribution')
axs[0, 1].set_xlabel('Annual Income')
axs[0, 1].set_ylabel('Frequency')

# Plotting Net Worth
axs[1, 0].hist(data['Net Worth'], bins=30, color='purple', edgecolor='black')
axs[1, 0].set_title('Net Worth Distribution')
axs[1, 0].set_xlabel('Net Worth')
axs[1, 0].set_ylabel('Frequency')

# Plotting Net Worth / Total Commitment
axs[1, 1].hist(data['Net Worth / Total Commitment'], bins=30, color='orange', edgecolor='black')
axs[1, 1].set_title('Net Worth / Total Commitment Distribution')
axs[1, 1].set_xlabel('Net Worth / Total Commitment')
axs[1, 1].set_ylabel('Frequency')

# Plotting Income / Total Commitment
axs[2, 0].hist(data['Income / Total Commitment'], bins=30, color='red', edgecolor='black')
axs[2, 0].set_title('Income / Total Commitment Distribution')
axs[2, 0].set_xlabel('Income / Total Commitment')
axs[2, 0].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Calculating the correlation matrix for the numerical columns
numerical_columns = data.select_dtypes(include=['number'])
correlation_matrix = numerical_columns.corr()

# Plotting the correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)
plt.title('Correlation Heatmap of Numerical Features')
plt.xticks(rotation=90, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Selecting wealth-related and fund-related boolean columns
wealth_columns = [col for col in data.columns if 'wealth -' in col]
fund_columns = [col for col in data.columns if 'fund -' in col]

# Calculating the counts of True values for wealth-related and fund-related columns
wealth_true_counts = data[wealth_columns].sum() / len(data)
fund_true_counts = data[fund_columns].sum() /len(data)

# Plotting bar charts for Wealth-related and Fund-related True values
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Wealth-related True values
axs[0].bar(wealth_true_counts.index, wealth_true_counts.values, color='blue', edgecolor='black')
axs[0].set_title('Wealth-related True Values')
axs[0].set_ylabel('Percentage')
axs[0].set_xticklabels(wealth_true_counts.index, rotation=45, ha='right')

# Fund-related True values
axs[1].bar(fund_true_counts.index, fund_true_counts.values, color='green', edgecolor='black')
axs[1].set_title('Fund-related True Values')
axs[1].set_ylabel('Percentage')
axs[1].set_xticklabels(fund_true_counts.index, rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# Plotting the count of occurrences for each industry in a bar chart
industry_counts = data['Industry'].value_counts()

# Creating the bar plot for industry counts
plt.figure(figsize=(10, 6))
plt.bar(industry_counts.index, industry_counts.values, color='purple', edgecolor='black')
plt.title('Industry Distribution')
plt.xlabel('Industry')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
