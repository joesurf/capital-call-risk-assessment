### Apply Red Flag Labelling

```python
df = pd.read_csv('../sample_data/cleaned_sample_data.csv')
```

```python
df.head()
```

```python
class Investor:
    INCOME: float
    FIXEDINCOME: float
    VARIABLEINCOME: float
    OTHERINCOME: float
    PERSONALASSETS: float
    RESIDENCEVALUE: float
    LIQUIDASSETS: float
    FINANCIALASSETS: float
    AGE: int
    HASLPA: bool
    HASCRITICALILLNESS: bool
    NUMBERSOURCEOFFUNDS: int
    COMMITMENTAMOUNT: float
    INCOMEQUALIFIED: bool
    PERSONALQUALIFIED: bool
    FINANCIALQUALIFIED: bool

```

```python
def red_flag_algorithm_based_on_qualifier(investor: Investor):
    num_red_flags = 0

    if investor.COMMITMENTAMOUNT == 0:
        raise Exception('Commitment Amount cannot be zero')


    # income qualifier
    if investor.INCOMEQUALIFIED:
        if investor.FIXEDINCOME:
            if investor.VARIABLEINCOME:
                if (investor.FIXEDINCOME + investor.VARIABLEINCOME) / investor.COMMITMENTAMOUNT < 2:
                    num_red_flags += 1

            else:
                if investor.FIXEDINCOME / investor.COMMITMENTAMOUNT < 1:
                    num_red_flags += 2


    # personal asset qualifier
    elif investor.PERSONALQUALIFIED:
        if investor.INCOME:
            if investor.LIQUIDASSETS:
                income_liquid_assets_to_ca_ratio = (investor.LIQUIDASSETS + investor.INCOME) / investor.COMMITMENTAMOUNT

                if income_liquid_assets_to_ca_ratio < 1:
                    num_red_flags += 3
                elif income_liquid_assets_to_ca_ratio < 2:
                    num_red_flags += 2

            else:
                if investor.INCOME / investor.COMMITMENTAMOUNT < 2:
                    num_red_flags += 1


    # financial asset qualifier
    elif investor.FINANCIALQUALIFIED:
        if investor.INCOME:
            if investor.FINANCIALASSETS:
                income_financial_assets_to_ca_ratio = (investor.FINANCIALASSETS + investor.INCOME) / investor.COMMITMENTAMOUNT

                if income_financial_assets_to_ca_ratio < 1:
                    num_red_flags += 3
                elif income_financial_assets_to_ca_ratio < 2:
                    num_red_flags += 2

            else:
                if investor.INCOME / investor.COMMITMENTAMOUNT < 2:
                    num_red_flags += 1


    qualifier_red_flags = num_red_flags


    # demographics
    if investor.AGE > 65:
        if investor.AGE > 75:
            if investor.HASLPA:
                num_red_flags += 1
            else:
                num_red_flags += 2
        
        if investor.INCOME / investor.COMMITMENTAMOUNT < 2:
            num_red_flags += 1

    if investor.HASCRITICALILLNESS:
        num_red_flags += 1

    if investor.NUMBERSOURCEOFFUNDS < 3:
        num_red_flags += 1


    demographic_red_flags = num_red_flags - qualifier_red_flags

    return num_red_flags
```

```python
def red_flag_algorithm_based_on_all_qualifier(investor: Investor):
    num_red_flags = 0

    if investor.COMMITMENTAMOUNT == 0:
        raise Exception('Commitment Amount cannot be zero')


    # income qualifier
    if investor.FIXEDINCOME:
        if investor.VARIABLEINCOME:
            if (investor.FIXEDINCOME + investor.VARIABLEINCOME) / investor.COMMITMENTAMOUNT < 2:
                num_red_flags += 1

        else:
            if investor.FIXEDINCOME / investor.COMMITMENTAMOUNT < 1:
                num_red_flags += 2


    # personal asset qualifier
    if investor.INCOME:
        if investor.LIQUIDASSETS:
            income_liquid_assets_to_ca_ratio = (investor.LIQUIDASSETS + investor.INCOME) / investor.COMMITMENTAMOUNT

            if income_liquid_assets_to_ca_ratio < 1:
                num_red_flags += 3
            elif income_liquid_assets_to_ca_ratio < 2:
                num_red_flags += 2

        else:
            if investor.INCOME / investor.COMMITMENTAMOUNT < 2:
                num_red_flags += 1


    # financial asset qualifier
    if investor.INCOME:
        if investor.FINANCIALASSETS:
            income_financial_assets_to_ca_ratio = (investor.FINANCIALASSETS + investor.INCOME) / investor.COMMITMENTAMOUNT

            if income_financial_assets_to_ca_ratio < 1:
                num_red_flags += 3
            elif income_financial_assets_to_ca_ratio < 2:
                num_red_flags += 2

        else:
            if investor.INCOME / investor.COMMITMENTAMOUNT < 2:
                num_red_flags += 1


    qualifier_red_flags = num_red_flags


    # demographics
    if investor.AGE > 65:
        if investor.AGE > 75:
            if investor.HASLPA:
                num_red_flags += 1
            else:
                num_red_flags += 2
        
        if investor.INCOME / investor.COMMITMENTAMOUNT < 2:
            num_red_flags += 1

    if investor.HASCRITICALILLNESS:
        num_red_flags += 1

    if investor.NUMBERSOURCEOFFUNDS < 3:
        num_red_flags += 1


    demographic_red_flags = num_red_flags - qualifier_red_flags

    return num_red_flags
```

```python
def red_flag_algorithm_based_on_each_qualifier(investor: Investor):
    income_red_flags = 0
    personal_red_flags = 0
    financial_red_flags = 0
    demographic_red_flags = 0

    if investor.COMMITMENTAMOUNT == 0:
        raise Exception('Commitment Amount cannot be zero')


    # income qualifier
    if investor.FIXEDINCOME:
        if investor.VARIABLEINCOME:
            if (investor.FIXEDINCOME + investor.VARIABLEINCOME) / investor.COMMITMENTAMOUNT < 2:
                personal_red_flags += 1

        else:
            if investor.FIXEDINCOME / investor.COMMITMENTAMOUNT < 1:
                personal_red_flags += 2


    # personal asset qualifier
    if investor.INCOME:
        if investor.LIQUIDASSETS:
            income_liquid_assets_to_ca_ratio = (investor.LIQUIDASSETS + investor.INCOME) / investor.COMMITMENTAMOUNT

            if income_liquid_assets_to_ca_ratio < 1:
                income_red_flags += 3
            elif income_liquid_assets_to_ca_ratio < 2:
                income_red_flags += 2

        else:
            if investor.INCOME / investor.COMMITMENTAMOUNT < 2:
                income_red_flags += 1


    # financial asset qualifier
    if investor.INCOME:
        if investor.FINANCIALASSETS:
            income_financial_assets_to_ca_ratio = (investor.FINANCIALASSETS + investor.INCOME) / investor.COMMITMENTAMOUNT

            if income_financial_assets_to_ca_ratio < 1:
                financial_red_flags += 3
            elif income_financial_assets_to_ca_ratio < 2:
                financial_red_flags += 2

        else:
            if investor.INCOME / investor.COMMITMENTAMOUNT < 2:
                financial_red_flags += 1


    # demographics
    if investor.AGE > 65:
        if investor.AGE > 75:
            if investor.HASLPA:
                demographic_red_flags += 1
            else:
                demographic_red_flags += 2
        
        if investor.INCOME / investor.COMMITMENTAMOUNT < 2:
            demographic_red_flags += 1

    if investor.HASCRITICALILLNESS:
        demographic_red_flags += 1

    if investor.NUMBERSOURCEOFFUNDS < 3:
        demographic_red_flags += 1


    return income_red_flags + demographic_red_flags, personal_red_flags + demographic_red_flags, financial_red_flags + demographic_red_flags
```

```python
for index, row in df.iterrows():
    investor = Investor()
    for key, value in row.to_dict().items():
        setattr(investor, key, value)
    redflag_onequalifier = red_flag_algorithm_based_on_qualifier(investor)
    redflag_allqualifier = red_flag_algorithm_based_on_all_qualifier(investor)
    income_red_flags, personal_red_flags, financial_red_flags = red_flag_algorithm_based_on_each_qualifier(investor)
    df.at[index, 'REDFLAGS_ONEQUALIFIER'] = redflag_onequalifier
    df.at[index, 'REDFLAGS_ALLQUALIFIER'] = redflag_allqualifier
    df.at[index, 'REDFLAGS_INCOMEQUALIFIER'] = income_red_flags
    df.at[index, 'REDFLAGS_PERSONALQUALIFIER'] = personal_red_flags
    df.at[index, 'REDFLAGS_FINANCIALQUALIFIER'] = financial_red_flags
 

```
